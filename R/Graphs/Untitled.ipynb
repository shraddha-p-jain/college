# Barplot

```r
x = c(2,4, 3, 2, 3,5, 6, 7, 10, 2, 5, 8, 8, 2, 3, 2, 5, 6, 2,3)
```

```r
library(repr)
options(repr.plot.height = 4, repr.plot.width= 5)
```

```r
barplot(x)
```

This just builts a barplot with the heights = x, Which is not what we wnat most of the times.

```r
barplot(table(x))
```

A more appropriate barplot.

### Manipulating the barplot

```r
barplot(table(x),width = c(4,7)) #This only works with different vector length.
barplot(table(x),width = c(7)) 
barplot(table(x),width = c(4)) # No difference in there 2
```

```r
a<-table(x)
```

```r
barplot(a, names.arg = c(LETTERS[1:8])) # naming the bars
abline(h=0)
```
